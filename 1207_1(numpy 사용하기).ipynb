```python
# numpy를 사용하지 않는 경우 !
import matplotlib.pyplot as plt

t = []
p2 = []
p3 = []

for i in range(0,50,2):
    t.append(i/10)
    p2.append(round((i/10)**2, 2))
    p3.append(round((i/10)**3, 2))
    
print(t)
print(p2)
print(p3)

plt.plot(t, t, 'r--', t, p2, 'bs', t, p3, 'g^')
plt.show()
```

```python
# numpy를 사용하는 경우 !
import numpy as np

t = np.arange(0., 5., 0.2)

plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```
