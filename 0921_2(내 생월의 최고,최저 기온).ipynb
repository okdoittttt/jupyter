```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')

data = csv.reader(f)

next(data)

high = []
low = []

for row in data:
    if row[-1] != '' and row[-2] != '':
        date = row[0].split('-')
        if 1996 <= int(date[0]):
            if date[1] == '04':
                high.append(float(row[-1]))
                low.append(float(row[-2]))
            

plt.title('Max temp, Min temp of my birth month')
plt.plot(high, color='red')
plt.plot(low, color='blue')
plt.show()
```
