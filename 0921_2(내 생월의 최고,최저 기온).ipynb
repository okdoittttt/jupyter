```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')

data = csv.reader(f)

next(data)

high = []
low = []
month = []

for row in data:
    if row[-1] != '' and row[-2] != '':
        date = row[0].split('-')
        if 1996 <= int(date[0]):
            if date[1] == '04':
                high.append(float(row[-1]))
                low.append(float(row[-2]))
                month.append(date[0])
            
print(month)
plt.figure(figsize=(20, 5))
plt.title('Max temp, Min temp of my birth month')
plt.plot(month, high, 'r--s', label='hight_temp')
plt.plot(month, low, 'b--s', label='low_temp')
plt.legend()
plt.show()
```
