```python
# 서울의 최고 기온, 최저 기온 출력
# 그래프 출력(100개의 데이터만)
# 총 데이터 개수 출력

import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', 'r', encoding='utf-8')

data = csv.reader(f, delimiter=',')

head = next(data)

max_temp = -999
max_date = ""
min_temp = 999
min_date = ""

temp_list = []
temp_date = []

count = 0
for row in data:
    count += 1
    if row[-1] == '' or row[-2] == '' or row[0] == '':
        row[-1] = 0
        row[-2] = 0
        row[0] = 0
    row[-1] = float(row[-1])
    row[-2] = float(row[-2])
    temp_list.append(row[-1])
    temp_date.append(row[0])
    if max_temp < row[-1]:
        max_date = row[0]
        max_temp = row[-1]
    elif min_temp > row[-2]:
        min_date = row[0]
        min_temp = row[-2]
        
# print(temp_date)
print('최고기온 : {}, 년월 : {}, 최저기온 : {}, 년월 : {}'.format(max_temp, max_date, min_temp, min_date))
print('데이터의 개수 : {}'.format(len(temp_list)))
```

```python
# 데이터 개수를 터미널에 출력, 그래프에 100개만 ploting 하도록 하라.
graph_list = [temp_list[i] for i in range(0, 6)]
graph_date = [temp_date[i] for i in range(0, 6)]
print(graph_list)

plt.figure(figsize=(20, 5))
plt.title('Max temp of 100 data')

# x 축 y 축
plt.plot(graph_date, graph_list, 'r--s')
plt.show()
```
