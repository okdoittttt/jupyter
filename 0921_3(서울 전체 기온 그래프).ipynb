```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', 'r', encoding='utf-8')

data = csv.reader(f, delimiter=',')

head = next(data)

max_temp = -999
max_date = ""
min_temp = 999
min_date = ""

temp_list = []

for row in data:
    if row[-1] == '' or row[-2] == '':
        row[-1] = 0
        row[-2] = 0
    row[-1] = float(row[-1])
    row[-2] = float(row[-2])
    temp_list.append(row[-1])
    if max_temp < row[-1]:
        max_date = row[0]
        max_temp = row[-1]
    elif min_temp > row[-2]:
        min_date = row[0]
        min_temp = row[-2]
        
print('최고기온 : {}, 년월 : {}, 최저기온 : {}, 년월 : {}'.format(max_temp, max_date, min_temp, min_date))
# print(temp_list)
```

```python
plt.title('Max temp')
plt.plot(temp_list)
plt.show()
```
