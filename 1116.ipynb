```python
import matplotlib.pyplot as plt

size = [10, 20, 40, 70]
label = ['A형', 'B형', 'C형', 'O형']
explode = [0, 0.10, 0, 0.0]

plt.title('파이 차트 그리기')
plt.rc('font', family='AppleGothic')
plt.pie(size,labels=label, autopct='%.1f%%', startangle=0, counterclock=False, explode=explode)
plt.axis('equal')
plt.legend()
plt.show()
```
