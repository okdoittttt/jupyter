```python
import csv
# import re

# file 오픈, 인코딩 설정
f = open('busan.csv', 'r', encoding='utf-8')
data = csv.reader(f, delimiter=',')

# 가장 첫 번째 위치한 데이터로 커서 이동
head = next(data)

# 데이터가 저장될 list
jingu = []
haeundae = []
youngdo = []
dongraegu = []

# 지역 이름이 해당하는 곳의 값을 변수에 저장하도록 한다.
for row in data:
    if '진구' in row[0]:
        for i in row[3:]:
            #데이터에 ','를 제거하는 작업, re 라이브러리를 사용해도 된다.
            i = i.replace(',','')
            # result = re.sub(r"[^0-9]", "", i)
            jingu.append(int(i))
    elif '해운대' in row[0]:
        for i in row[3:]:
            i = i.replace(',','')
            haeundae.append(int(i))
    elif '영도' in row[0]:
        for i in row[3:]:
            i = i.replace(',','')
            youngdo.append(int(i))
    elif '동래' in row[0]:
        for i in row[3:]:
            i = i.replace(',','')
            dongraegu.append(int(i))

        
# 값 확인
print("=============== 진 구 ==================")
print(jingu)
print("=============== 해 운 대 구 ==================")
print(haeundae)
print("=============== 영 도 구 ==================")
print(youngdo)
print("=============== 동 래 구 ==================")
print(dongraegu)
```

```python
import matplotlib.pyplot as plt

# 각 데이터들을 그래프에 출력한다.
plt.rc('font', family='AppleGothic')# 한글출력을 위한 전처리
plt.title("부산 구별 인구데이터 분석")# 타이틀 설정
# 각 데이터 그래프 출력하기
plt.plot(haeundae, 'r--s', label='해운대구')
plt.plot(youngdo, 'y--*', label='영도구')
plt.plot(jingu, 'b--o', label='진구')
plt.plot(dongraegu, 'g--d', label='동래구')
plt.legend()
# 실행
plt.show()
```
