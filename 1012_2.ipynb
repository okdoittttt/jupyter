```python
import csv
f = open('seoul_day.csv', encoding='utf-8')
data = csv.reader(f)
next(data)

result = []
day = []

for row in data :
    if row[-1] != '' :
        result.append(float(row[-1]))
        day.append(row[0])
        
print(day[0])
print(result)

import matplotlib.pyplot as plt

#figsize(가로길이,세로길이)
plt.figure(figsize = (10,2), dpi = 600)
plt.title('max temp')
plt.plot(day[0:20], result[0:20], 'r^-', label='temp')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv
f = open('seoul_day_all.csv', encoding='utf-8')
data = csv.reader(f)
next(data)

high = [] # row[-1]  max temp
low = []  # row[-2]  min temp

for row in data :  # row[0] : date
    if row[-1] != '' :
        if row[0].split('.')[1] == '2':  # month
            high.append(float(row[-1]))
            
    if row[-2] != '' :
        if row[0].split('.')[1] == '2':
            low.append(float(row[-2]))

# sLow = sum(low) / len(low)
# print('Low Temp')
# print('max =', max(low))
# print('min =', min(low))
# print('ave =', round(sLow,2))
plt.title('Low Temp')
plt.hist(low, bins = 1000, color = 'b')
plt.show()
            
# print('High Temp')
# sHigh = sum(high) / len(high)
# print('max =', max(high))
# print('min =', min(high))
# print('ave =', round(sHigh,2))
plt.title('High Temp')
plt.hist(high, bins = 1000, color = 'r')
plt.show()
```

```python
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug = []

for row in data :
    month = row[0].split('.')[1]
    day = row[0].split('.')[2]
    # print(month, day)
    if row[-1] != '' :
        if month == '8' and day == '5':
            aug.append(float(row[-1]))

import matplotlib.pyplot as plt
#plt.figure(dpi = 300)
plt.rc('font', family = 'AppleGothic') # 맥사용자
plt.title('8/5의 최고 기온')
plt.hist(aug, bins = 100, color = 'r', label = '8/5')
plt.legend()
plt.show()
```

```python
import csv
f = open('seoul.csv', encoding='utf-8')
data = csv.reader(f)
next(data)
aug = []
jan = []

for row in data :
    month = row[0].split('.')[1]
    day = row[0].split('.')[2]
    if row[-1] != '' :
        if month == '8' and day == '12':
            aug.append(float(row[-1]))
        if month == '1'and day == '12':
            jan.append(float(row[-1]))
print(len(aug))

import matplotlib.pyplot as plt
plt.rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

plt.title('8/12과 1/2의 최고 기온')
plt.hist(aug, bins = 100, color = 'r', label = '8/12')
plt.hist(jan, bins = 100, color = 'b', label = '1/12')
plt.legend()
plt.show()
```

## boxplot
- 최소값
- 제 1사분위 수 (Q1)
- 제 2사분위 수 또는 중위수 (Q2)
- 제 3사분위 수 (Q3)
- 최대값

```python
import matplotlib.pyplot as plt
#plt.figure(dpi = 300)
import random
result = []
for i in range(13) :
    result.append(random.randint(1,1000))
print(sorted(result))

import numpy as np
result = np.array(result)
print("1/4: " + str(np.percentile(result,25)))
print("2/4: " + str(np.percentile(result,50)))
print("3/4: " + str(np.percentile(result,75)))

plt.boxplot(result)
plt.show()
```
