```python
import csv

f = open('agemf.csv', 'r', encoding='utf-8')
data = csv.reader(f, delimiter=',')
head = next(data)

location = input('지역을 입력하세요.')

man = []
woman = []

result_man = []
result_woman = []

for row in data:
    if location in row[0]:
        for i in row[3:104]:
            m_data = int(i.replace(',',''))
            man.append(m_data)
            result_man = sum(man)
        for i in row[106:]:
            w_data = int(i.replace(',',''))
            woman.append(w_data)
            result_woman = sum(woman)
        break
        
print(result_man)
print(result_woman)

result = []
result.append(int(result_man))
result.append(int(result_woman))

print(result)
```

```python
import matplotlib.pyplot as plt

label = ['남자', '여자']
ex = [0.0, 0.10]
colors = ['silver', 'gold']

plt.title('파이 차트 그리기')
plt.rc('font', family='AppleGothic')
plt.pie(result,labels=label, autopct='%.1f%%', startangle=90, counterclock=False, explode=ex, colors=colors)
plt.axis('equal')
plt.legend()
plt.show()
```
