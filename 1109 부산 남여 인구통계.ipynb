```python
import csv

f = open('agemf.csv', 'r', encoding='utf-8')
data = csv.reader(f, delimiter=',')
head = next(data)

location = input("지역을 입력하세요.")

man = []
woman = []

# version 1
# for row in data:
#     if location in row[0]:
#         for i in row[3:104]:
#             m_data = int(i.replace(',',''))
#             man.append(m_data*-1)
#         for i in row[106:]:
#             w_data = int(i.replace(',',''))
#             woman.append(w_data)

# version 2
# for row in data:
#     if '부산진구' in row[0]:
#         for i in range(0, 101):
#             m_data = int(row[i+3].replace(',',''))
#             w_data = int(row[-(i+1)].replace(',',''))
#             man.append(m_data)
#             woman.append(w_data)

# woman 데이터 재배치
# woman.reverse()

# version 3(부산시 전체)
for row in data:
    if location in row[0]:
        for i in row[3:104]:
            m_data = int(i.replace(',',''))
            man.append(m_data*-1)
        for i in row[106:]:
            w_data = int(i.replace(',',''))
            woman.append(w_data)
        break

print('============== {}의 남자인구 / len : {} ================='.format(location, len(man)))
print(man)
print('============== {}의 여자인구 / len : {} ================='.format(location, len(woman)))
print(woman)
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family='AppleGothic')# 한글출력을 위한 전처리
plt.rcParams['axes.unicode_minus'] = False # '-'(마이너스) 출력을 위한 전처리
plt.title('부산 지역별 남여 인구 통계 그래프')

# 그레프 출력
plt.barh(range(101), man, color='r', label='남자')
plt.barh(range(101), woman, color='b', label='여자')

plt.legend()#범례
plt.show()
```
