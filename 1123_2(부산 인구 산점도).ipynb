```python
import csv
import matplotlib.pyplot as plt

# 읽을 file을 받아와 데이터 분석 준비
f = open('agemf.csv', 'r', encoding='utf-8')
data = csv.reader(f, delimiter=',')
head = next(data)

# 검색할 인구를 받는다.
location = input("지역을 입력하세요.")

# 남, 녀 인구수 입력받을 리스트
man = []
woman = []

# 검색한 지역명에 남, 녀 인구수 입력받기
for row in data:
    if location in row[0]:
        # 남자
        for i in row[3:104]:
            m_data = int(i.replace(',',''))
            man.append(m_data)
        # 여자
        for i in row[106:]:
            w_data = int(i.replace(',',''))
            woman.append(w_data)
        # '부산광역시' 검색시 사용될 break문
        break

print('============== {}의 남자인구 / len : {} ================='.format(location, len(man)))
print(man)
print('============== {}의 여자인구 / len : {} ================='.format(location, len(woman)))
print(woman)
```

```python
plt.rc('font', family='AppleGothic')
plt.title('{} 남녀 인구 분포 산점도 그래프'.format(location))

# scatter => 산점도 그래프
# x, y축 입력 | c => 색상, range =>색상 개수, cmap => 색상선택
# xlabel, ylabel => 축 이름
plt.scatter(man, woman, c=range(101), cmap='jet')
plt.colorbar()
plt.plot(range(max(man)), range(max(man)), 'g')
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')

plt.show()
```
