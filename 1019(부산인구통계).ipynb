```python
import csv
# import re

f = open('busan.csv', 'r', encoding='utf-8')
data = csv.reader(f, delimiter=',')
head = next(data)

name = input('원하는 지역을 입력하세요.')

population = []

for row in data:
    if name in row[0]:
        for i in row[3:]:
            i = i.replace(',','')
            # result = re.sub(r"[^0-9]", "", i)
            population.append(int(i))

print(population)
```

```python
import matplotlib.pyplot as plt

plt.plot(population)
plt.show()
```
