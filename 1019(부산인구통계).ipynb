```python
import csv
# import re

f = open('busan.csv', 'r', encoding='utf-8')
data = csv.reader(f, delimiter=',')
head = next(data)

name = input('원하는 지역을 입력하세요.')

population = []

for row in data:
    if name in row[0]:
        for i in row[3:]:
            i = i.replace(',','')
            # result = re.sub(r"[^0-9]", "", i)
            population.append(int(i))

print(population)
```

```python
import matplotlib.pyplot as plt

# 각 데이터들을 그래프에 출력한다.
plt.rc('font', family='AppleGothic')# 한글출력을 위한 전처리
plt.title("부산 구별 인구데이터 분석")# 타이틀 설정
# 각 데이터 그래프 출력하기

# 선 그래프
# plt.plot(population, 'r--s', label='인구')

# 막대 그래프
plt.bar(range(101), population, color='r')

plt.legend()
# 실행
plt.show()
```
