```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,30,20,40]

# color = ['red', 'blue', 'yellow', 'green']
size = [100, 300, 200, 600]

plt.rc('font', family='AppleGothic')
plt.title('산점도 찍어보기')

plt.scatter(x,y, c=range(4), cmap='jet', s=size)
plt.colorbar()
plt.xlabel('남자')
plt.ylabel('여자')

plt.show()
```
